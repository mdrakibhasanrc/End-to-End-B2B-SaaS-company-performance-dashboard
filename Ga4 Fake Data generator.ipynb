## Data Collection:
For this Project I am Creating Fake Dataset using Python. 

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set the random seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 3000

# Generate Website Data
website_data = {
    'website_id': range(1, num_samples + 1),  # Unique ID for each website entry
    'event_date': [datetime.now() - timedelta(days=random.randint(0, 30)) for _ in range(num_samples)],
    'session': [f'session_{i}' for i in range(1, num_samples + 1)],
    'event_name': random.choices(['page_view', 'sign_up', 'purchase', 'login', 'logout'], k=num_samples),
    'user_id': [f'user_{random.randint(1, 500)}' for _ in range(num_samples)],
    'page_location': [f'www.dataology.com/{random.choice(["home", "sign_up", "login", "dashboard", "product"])}' for _ in range(num_samples)],
    'website_visits': np.random.randint(1000, 10000, num_samples),
    'bounce_rate': np.random.uniform(0.1, 0.5, num_samples),
    'avg_session_duration': np.random.uniform(30, 300, num_samples),
    'unique_visitors': np.random.randint(500, 5000, num_samples),
    'conversion_rate': np.random.uniform(0.01, 0.2, num_samples),
    'page_views': np.random.randint(2000, 20000, num_samples),
    'traffic_source': random.choices(['Organic', 'Paid', 'Social', 'Direct'], k=num_samples),
    'device_category': random.choices(['Mobile', 'Desktop', 'Tablet'], k=num_samples),
    'geographic_region': random.choices(['North America', 'Europe', 'Asia'], k=num_samples),
    'conversion': np.random.randint(0, 100, num_samples)  # New conversion column
}

# Generate Marketing Data
marketing_data = {
    'campaign_id': range(1, num_samples + 1),  # Unique ID for each marketing campaign
    'campaign_name': [f'Campaign {i}' for i in range(1, num_samples + 1)],
    'ad_spend': np.random.randint(1000, 20000, num_samples),
    'leads_generated': np.random.randint(1, 500, num_samples),
    'click_through_rate': np.random.uniform(0.01, 0.15, num_samples),
    'impressions': np.random.randint(1000, 50000, num_samples),
    'source': random.choices(['Email', 'Social Media', 'Referral', 'Direct'], k=num_samples),
    'medium': random.choices(['CPC', 'Organic', 'Email', 'Social'], k=num_samples),
    'start_date': [datetime.now() - timedelta(days=random.randint(1, 30)) for _ in range(num_samples)],
    'end_date': [datetime.now() + timedelta(days=random.randint(1, 30)) for _ in range(num_samples)],
    'click': np.random.randint(0, 1000, num_samples)  # New click column
}

# Generate Sales Data
sales_data = {
    'sales_rep_id': range(1, num_samples + 1),  # Unique ID for each sales representative entry
    'sales_rep_name': [f'Rep {i}' for i in range(1, num_samples + 1)],
    'deals_closed': np.random.randint(0, 100, num_samples),
    'revenue_generated': np.random.uniform(5000, 500000, num_samples),  # in dollars
    'average_deal_size': np.random.uniform(500, 20000, num_samples),
    'sales_cycle_length': np.random.randint(5, 90, num_samples),  # in days
    'new_customers': np.random.randint(0, 50, num_samples),
    'churn_rate': np.random.uniform(0.0, 0.2, num_samples),
    'upsell_revenue': np.random.uniform(1000, 20000, num_samples),
    'customer_lifetime_value': np.random.uniform(1000, 100000, num_samples),
    'date': [datetime.now() - timedelta(days=random.randint(0, 365)) for _ in range(num_samples)]  # Random date within one year
}

# Generate Customer Satisfaction Data
customer_satisfaction_data = {
    'customer_id': range(1, num_samples + 1),  # Unique ID for each customer
    'satisfaction_score': np.random.randint(1, 11, num_samples),
    'net_promoter_score': np.random.randint(-100, 100, num_samples),
    'feedback': [f'Feedback {i}' for i in range(1, num_samples + 1)],
    'support_tickets': np.random.randint(0, 20, num_samples),
    'resolution_time': np.random.randint(1, 48, num_samples),  # in hours
    'repeat_purchase': random.choices(['Yes', 'No'], k=num_samples),
    'last_purchase_date': [datetime.now() - timedelta(days=np.random.randint(1, 365)) for _ in range(num_samples)],
    'customer_segment': random.choices(['Small Business', 'Mid-Market', 'Enterprise'], k=num_samples),
    'referral_source': random.choices(['Social Media', 'Word of Mouth', 'Email', 'Ads'], k=num_samples)
}

# Create DataFrames
website_df = pd.DataFrame(website_data)
marketing_df = pd.DataFrame(marketing_data)
sales_df = pd.DataFrame(sales_data)
customer_satisfaction_df = pd.DataFrame(customer_satisfaction_data)

```
